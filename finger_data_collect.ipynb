```python
import cv2
import mediapipe as mp
import time
import csv

# 미디어파이프 손 모델 초기화
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(max_num_hands=2)  # 최대 2개의 손 감지

# OpenCV 비디오 캡처 초기화 (웹캠 사용)
cap = cv2.VideoCapture(0)

# 1초에 1번 찍기 위한 설정
fps = 1
interval = int(1 / fps)

# 25초 동안 실행
for second in range(100):
    # 현재 시간 기록
    start_time = time.time()

    # 손 위치 확인 및 좌표 저장
    landmarks_list_left = []
    landmarks_list_right = []

    # 1초에 1번씩 찍은 21개의 랜드마크 좌표를 CSV 파일로 저장
    for frame_number in range(fps):
        # 프레임 읽기
        ret, frame = cap.read()
        if not ret:
            print("프레임 읽기 오류")
            break

        # 손 감지 수행
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = hands.process(frame_rgb)

        if results.multi_hand_landmarks:
            for i, landmarks in enumerate(results.multi_hand_landmarks):
                landmarks_count = len(landmarks.landmark)
                landmarks_count_limit = 21  # 개별 손마다 최대 21개의 랜드마크만 사용
                if landmarks_count > landmarks_count_limit:
                    continue

                for j, point in enumerate(landmarks.landmark):
                    x, y = int(point.x * frame.shape[1]), int(point.y * frame.shape[0])
                    if j < landmarks_count_limit:
                        if i == 0:
                            landmarks_list_right.append([x, y])
                        elif i == 1:
                            landmarks_list_left.append([x, y])

        # 화면에 출력
        cv2.imshow('손 랜드마크 검출', frame)

        # 'q' 키를 누르면 종료
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

        # 1초 대기
        time.sleep(interval)

    # 1초 쉬기
    time.sleep(1)

    # CSV 파일로 저장
    filename = f'landmarks_2_{second + 301}.csv'
    with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
        csv_writer = csv.writer(csvfile)
        csv_writer.writerow(['x_r', 'y_r', 'x_l', 'y_l'])

        max_len = max(len(landmarks_list_right), len(landmarks_list_left))
        for i in range(max_len):
            right = landmarks_list_right[i] if i < len(landmarks_list_right) else ["0", "0"]
            left = landmarks_list_left[i] if i < len(landmarks_list_left) else ["0", "0"]
            csv_writer.writerow(right + left)

# 종료
cap.release()
cv2.destroyAllWindows()
```

```python
#각 손의 랜드마크에서 최소 경계 사각형을 계산하고, 해당 사각형 내의 좌표만을 추출
import cv2
import mediapipe as mp
import time
import csv
import numpy as np

# 미디어파이프 손 모델 초기화
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(max_num_hands=2)  # 최대 2개의 손 감지

# OpenCV 비디오 캡처 초기화 (웹캠 사용)
cap = cv2.VideoCapture(0)

# 1초에 1번 찍기 위한 설정
fps = 1
interval = int(1 / fps)

# 최소 경계 사각형 계산 함수
def calculate_bounding_box(landmarks):
    x_values = [landmark[0] for landmark in landmarks]
    y_values = [landmark[1] for landmark in landmarks]
    min_x = min(x_values)
    max_x = max(x_values)
    min_y = min(y_values)
    max_y = max(y_values)
    return min_x, min_y, max_x, max_y

# 25초 동안 실행
for second in range(50):
    # 현재 시간 기록
    start_time = time.time()

    # 손 위치 확인 및 좌표 저장
    landmarks_list_left = []
    landmarks_list_right = []

    # 1초에 1번씩 찍은 21개의 랜드마크 좌표를 CSV 파일로 저장
    for frame_number in range(fps):
        # 프레임 읽기
        ret, frame = cap.read()
        if not ret:
            print("프레임 읽기 오류")
            break

        # 손 감지 수행
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = hands.process(frame_rgb)

        if results.multi_hand_landmarks:
            for i, landmarks_per_hand in enumerate(results.multi_hand_landmarks):
                landmarks = np.array([[landmark.x * frame.shape[1], landmark.y * frame.shape[0]] for landmark in landmarks_per_hand.landmark])

                # 최소 경계 사각형 계산
                min_x, min_y, max_x, max_y = calculate_bounding_box(landmarks)

                # 최소 경계 사각형 내의 랜드마크 좌표 추출
                landmarks_in_bbox = []
                for landmark in landmarks:
                    if min_x <= landmark[0] <= max_x and min_y <= landmark[1] <= max_y:
                        landmarks_in_bbox.append(landmark)

                # 최소 경계 사각형 내의 좌표를 CSV 파일에 저장
                if i == 0:
                    landmarks_list_right.append(landmarks_in_bbox)
                elif i == 1:
                    landmarks_list_left.append(landmarks_in_bbox)

        # 화면에 출력
        cv2.imshow('손 랜드마크 검출', frame)

        # 'q' 키를 누르면 종료
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

        # 1초 대기
        time.sleep(interval)

    # 1초 쉬기
    time.sleep(1)

    # CSV 파일로 저장
    filename = f'landmarks_0_{second + 0}.csv'
    with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
        csv_writer = csv.writer(csvfile)
        csv_writer.writerow(['x_r', 'y_r', 'x_l', 'y_l'])

        max_len = max(len(landmarks_list_right), len(landmarks_list_left))
        for i in range(max_len):
            right = landmarks_list_right[i] if i < len(landmarks_list_right) else []
            left = landmarks_list_left[i] if i < len(landmarks_list_left) else []

            # 최소 21개의 랜드마크가 저장되도록 0으로 패딩
            right.extend([[0, 0]] * (21 - len(right)))
            left.extend([[0, 0]] * (21 - len(left)))

            # 한 행에 오른손의 랜드마크 좌표, 왼손의 랜드마크 좌표 저장
            csv_writer.writerow([coord for point in right for coord in point] + [coord for point in left for coord in point])

# 종료
cap.release()
cv2.destroyAllWindows()
```

```python
#좌표 정규화해서 저장하는 코드

import cv2
import mediapipe as mp
import numpy as np
import time
import csv

# 손 모델 로딩
mp_hands = mp.solutions.hands
hands = mp_hands.Hands()

# 웹캠 열기
cap = cv2.VideoCapture(0)

# 반복할 시간 설정 (초)
duration = 7

# 프로그램 시작 시간 기록
start_time = time.time()

# 카운터 초기화
capture_count = 880

# 반복 실행
while (time.time() - start_time) < duration:
    # 프레임 캡처
    ret, frame = cap.read()
    if not ret:
        print("Failed to capture frame from webcam.")
        exit()

    # OpenCV를 사용하여 BGR 이미지를 RGB로 변환
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Mediapipe를 사용하여 양손 감지
    results = hands.process(rgb_frame)

    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            # 감지된 손의 landmark를 사용하여 bounding box 좌표 계산
            landmarks = hand_landmarks.landmark
            x_coords = [landmark.x for landmark in landmarks]
            y_coords = [landmark.y for landmark in landmarks]

            # Calculate bounding box size
            x_min, y_min = min(x_coords), min(y_coords)
            x_max, y_max = max(x_coords), max(y_coords)
            box_width = x_max - x_min
            box_height = y_max - y_min

            # Check if the hand is left or right
            if landmarks[mp_hands.HandLandmark.WRIST].x < 0.5:
                hand_type = "left"
            else:
                hand_type = "right"

            # Calculate bounding box size
            x_min, y_min = min(x_coords), min(y_coords)
            x_coords_shifted = [x - x_min for x in x_coords]
            y_coords_shifted = [y - y_min for y in y_coords]

            # Normalize coordinates by box width and height
            x_coords_normalized = [x * 100 / box_width for x in x_coords_shifted]
            y_coords_normalized = [y * 100 / box_height for y in y_coords_shifted]

            # Save the normalized coordinates to a CSV file
            csv_file_path = f"{hand_type}_landmarks_etc_{capture_count}.csv"
            with open(csv_file_path, 'w', newline='') as csv_file:
                csv_writer = csv.writer(csv_file)
                csv_writer.writerow(['X', 'Y'])
                for x, y in zip(x_coords_normalized, y_coords_normalized):
                    csv_writer.writerow([x, y])

    # Display the frame
    cv2.imshow("Hand Detection", frame)

    # Increase the capture count
    capture_count += 1

    # Wait for 1 millisecond
    cv2.waitKey(1)

# 사용된 자원 해제
cap.release()
cv2.destroyAllWindows()
```
